# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus. É uma doença respiratória aguda causada por um vírus chamado SARS-CoV-2. A pandemia de COVID-19 começou na China em dezembro de 2019 e se espalhou rapidamente pelo mundo.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

<center><img alt="Colaboratory logo" width="50%" src="https://cdn.pixabay.com/photo/2020/03/19/21/35/covid-4948866_960_720.jpg"></center>

No Brasil, o primeiro caso de COVID-19 foi confirmado em 26 de fevereiro de 2020. Desde então, o número de casos e de mortes no país tem aumentado rapidamente. 

O governo federal e os governos estaduais tomou medidas para tentar conter a disseminação da doença, incluindo a implementação de medidas de distanciamento social, o fechamento de escolas e empresas e o uso de máscaras em lugares públicos.

Apesar dessas medidas, o número de casos e de mortes em dezembro de 2020, o Brasil ultrapassou a marca de 7 milhões de casos confirmados e aproximadamente 180 mil mortes. 

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## **Obtenção dos Dados**

Para esse projeto vamos utilizar os dados da *Our World in Data* que pode ser consultado através [deste link.](https://github.com/owid/covid-19-data/tree/master/public/data)

**Our World in Data** é uma organização sem fins lucrativos que coleta e analisa dados sobre uma variedade de questões globais, incluindo saúde, economia e meio ambiente. O *dataset sobre o COVID-19 (coronavirus) da Our World in Data* inclui informações sobre o número de casos e de mortes causados ​​pelo COVID-19 em todo o mundo, assim como dados demográficos, como idade e gênero das pessoas afetadas.

Os dados são coletados de várias fontes, incluindo governos, autoridades de saúde e organizações internacionais, como a Organização Mundial da Saúde (OMS). A Our World in Data também realiza uma análise cuidadosa dos dados para garantir que eles sejam precisos e atualizados.

Os usuários do dataset sobre o COVID-19 da Our World in Data podem esperar encontrar dados confiáveis ​​e atualizados sobre a pandemia de COVID-19 em todo o mundo. Os dados podem ser usados ​​para entender como a doença está se espalhando, quais grupos de pessoas estão sendo mais afetados e como os governos e as autoridades de saúde estão lidando com a pandemia.

### **Dificuldade de Análise dos Dados**

Analisar uma doença nova, como o COVID-19, pode ser particularmente desafiador por várias razões. Uma dificuldade é que muito pouco se sabe sobre a doença no início de uma pandemia. Isso pode dificultar a compreensão dos sintomas, da transmissão e do tratamento da doença. Além disso, o aparecimento de uma doença nova pode gerar muita incerteza e medo na população, o que pode levar a comportamentos de pânico e descontrole.

Outra dificuldade é que os dados e informações sobre a doença podem ser conflitantes ou incompletos. Isso pode ser especialmente verdadeiro durante os primeiros meses de uma pandemia, quando os governos e as autoridades de saúde estão ainda tentando entender a doença e estabelecer protocolos de resposta.

Além disso, os governos e as autoridades de saúde podem ter dificuldade em coletar e compilar dados precisos e completos sobre a doença. Isso pode ser devido a problemas de capacidade ou ao fato de que algumas informações podem ser confidenciais ou sensíveis.

Outro problema é que os dados coletados pelos governos podem não ser compreensíveis ou acessíveis para o público em geral. Isso pode levar a especulações e boatos sobre a doença e a disseminação de informações incorretas ou enganosas.

Por fim, os dados coletados pelos governos podem não ser suficientemente detalhados ou específicos para permitir uma análise aprofundada da doença. Isso pode tornar mais difícil para os cientistas e os profissionais de saúde compreenderem a doença e encontrar maneiras de tratá-la e controlá-la.

* `tests_units` - Unidades utilizadas pelo local para relatar seus dados de teste.
* `total_vaccinations` - Total de doses administradas da vacina contra a COVID-19.
* `people_vaccinated` - Total de pessoas que receberam ao menos uma dose da vacina contra a COVID-19.
* `people_fully_vaccinated` - Total de pessoas que estão completamente vacinadas (duas doses ou não) contra a COVID-19.
* `new_vaccinations` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos).
* `new_vaccinations_smoothed` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos) nos últimos 7 dias.
* `total_vaccinations_per_hundred` - Novas doses de vacinas administradas contra a COVID-19 (apenas calculado para dias consecutivos) por cem pessoas.
* `people_vaccinated_per_hundred` - Total de pessoas que receberam ao menos uma dose da vacina contra a COVID-19 a cada cem pessoas.
* `people_fully_vaccinated_per_hundred` - Total de pessoas que estão completamente vacinadas (duas doses ou não) contra a COVID-19 a cada cem pessoas.

* `stringency_index` - Índice de Restrição de Resposta Governamental: medida composta baseada em 9 indicadores de resposta incluindo encerramentos de escolas, encerramentos de locais de trabalho e proibições de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita).
* `population` - População (última atualização).

* `median_age` - Média de idade da população, de acordo com a projeção da ONU de 2020.
* `aged_65_older` - Parcela da população com 65 anos ou mais, de acordo com o ano mais recente disponível.
* `aged_70_older` - Parcela da população com 70 anos ou mais, de acordo com os dados de 2015.
* `gdp_per_capita` - Produto Interno Bruto (PIB) per capita.
* `extreme_poverty` - Parcela da população que vive em extrema pobreza, de acordo com o ano mais recente disponível a partir de 2010.
* `cardiovasc_death_rate` - Taxa de morte por doenças cardiovasculares, de acordo com dados de 2017 (número anual de mortes por 100 mil pessoas).
* `diabetes_prevalence` - Prevalência de diabetes (% da população de 20 a 79 anos), de acordo com dados de 2017.
* `female_smokers` - Parcela da população de mulheres fumantes, de acordo com o ano mais recente disponível.
* `male_smokers` - Parcela da população de homens fumantes, de acordo com o ano mais recente disponível.
* `handwashing_facilities` - Parcela da população com instalações básicas para lavagem das mãos no local, de acordo com o ano mais recente disponível.
* `hospital_beds_per_thousand` - Leitos em hospitais por mil pessoas, de acordo com o ano mais recente disponível desde 2010.
* `life_expectancy` - Expectativa de vida ao nascer em 2019.
* `human_development_index` - Índice de desenvolvimento humano: Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano, são eles: uma vida longa e saudável, conhecimento e um padrão de vida decente.
* `excess_mortality` - Excesso de mortalidade para todas as idades.

```
# vamos localizar as entradas associadas à data atual e agrupá-las pelo número total de mortes
df.loc[df.date == '2022-12-20'].sort_values(by = 'total_deaths', ascending = False).head(10)
```

Podemos perceber que: em nossa coluna `location` temos algumas entradas que não são países, mas sim regiões/setores geográficos. Por conta disso, vamos criar um *dataset* auxiliar a partir do original que contenha apenas países, viabilizando uma visualização gráfica associada aos países do nosso conjunto de dados.

```
# criando uma cópia
df_paises = df.copy()

# eliminando linhas que apresentam valores ausentes na coluna continent 
# (que correspondem  às regiões/setores mencionados acima).
df_paises = df.dropna(subset = ['continent'], axis = 0)

# novamente, vamos localizar as entradas associadas à data atual e agrupá-las pelo número total de mortes
top_10_deaths = df_paises.loc[df_paises.date == '2022-12-20', ['location','total_deaths']].sort_values(by = 'total_deaths', ascending = False).head(10)

# visualização tabular dos 10 países com mais casos de morte 
top_10_deaths
```

### **Top 10 países com mais casos de morte até o momento**

Para essa demostração, vamos utilizar o gráfico de barras, pois é uma ferramenta útil para visualizar e comparar dados categóricos ou numéricos de maneira clara e concisa. 

Eles permitem comparações fáceis, mostram como os dados mudam ao longo do tempo e permitem que os usuários vejam rapidamente como diferentes categorias ou grupos se comparam em relação a um determinado conjunto de dados. 

Isso pode ser útil para identificar padrões ou tendências nos dados.

```
# criando objetos da biblioteca Matplotlib
fig, ax = plt.subplots(figsize=(13,6))

# plotando gráfico de barras
top_10_deaths.plot(x="location", y="total_deaths", kind="bar", ax=ax, label='Total de Mortes')

ax.set_title("Top 10 países com mais mortes por COVID-19")
ax.set_xlabel("Países")
ax.set_ylabel("Número de mortes")

plt.tight_layout()
```

Agora sim conseguimos ter uma melhor visualização! E também não podemos deixar de notar que (infelizmente) o Brasil ocupa o segundo lugar no gráfico acima. Mas antes de focar nossa análise no escopo brasileiro , vamos para mais uma pergunta investigativa.

```
# realizando nossa seleção de dados 
brasil_total_cases = df.loc[df.location == 'Brazil', ['total_cases', 'date']]

# criando objetos da biblioteca Matplotlib
fig, ax = plt.subplots(figsize=(13,6))

# plotando gráfico
brasil_total_cases.plot(kind = 'line', x = 'date', y = 'total_cases',label='Total de Casos no Brasil', logx = False, logy = True, ax = ax)

ax.set_title("Casos Confirmados no Brasil")
ax.set_xlabel("Data")
ax.set_ylabel("Total de casos")

plt.tight_layout()
```

Pudemos observar que, assim como ocorreu no mundo, no Brasil houve um grande crescimento percentual, no número de casos, a partir de meados de Janeiro de 2020, seguido de um período de estabilização, depois com um suave crescimento no final de 2021 e finalmente um novo período de estabilização.
