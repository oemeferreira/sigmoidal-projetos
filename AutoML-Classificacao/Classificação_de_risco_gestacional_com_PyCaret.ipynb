## **Preparação dos dados**

A variavel alvo (`fetal_health`) está como `float64`, o que significa que ela está armazenando valores com casas decimais. Isso é um problema, pois algoritmos de classificação esperam valores inteiros para serem utilizados como rótulos de classes. Além disso, esses valores representam categorias (saudável, suspeito de patologia ou com patologia confirmada) e não valores numéricos com casas decimais.

Para tratar esse problema, precisamos converter a coluna fetal_health para o tipo `int64`, de forma que os valores 1, 2, 3 sejam convertidos para 0, 1, 2 respectivamente. Isso garantirá que nossa variável alvo esteja no formato correto para ser utilizada em algoritmos de classificação do PyCaret.

```
#criando uma cópia do dataframe
df_copy = df

#transformando de float para int
df_copy['fetal_health'] = df_copy['fetal_health'].astype("int64")

#Verificando o tipo da variável alvo
print(df_copy.dtypes['fetal_health'])

#Substituir valores
df_copy["fetal_health"] = df_copy["fetal_health"].replace([1, 2, 3],[0, 1, 2])

df_copy["fetal_health"].unique()
```

### **Comparando Modelos**

Nesta etapa, vamos comparar diferentes modelos de machine learning para avaliar sua performance em nosso conjunto de dados.

Existem diversas métricas e ferramentas que auxiliam na avaliação de um modelo de Machine Learning. Aqui, usaremos como ferramentas a *Confusion Matrix* e o *Classification Report* e como métricas utilizaremos principalmente a *Acurácia*, a *AUC* e o *Recall*.

De maneira resumida, o funcionamento dessas ferramentas e métricas é o seguinte:

* **Confusion Matrix:** plota uma matriz que possui, eu sei eixo Y, os valores verdadeiros (originais) dos dados de teste e no eixo X os valores previstos pelo algoritmo para cada classificação possível da variável alvo;
* **Classification Report:** relatório com as principais métricas utilizadas divididas para cada classificação possível da variável alvo;
* **Acurácia:** fração de predições corretas (no total). Pode ser resumida em "Quanto o modelo está acertando?". É preciso tomar cuidado com essa métrica, pois pode nos dar uma falsa sensação de bom desempenho, pois um modelo pode ter uma acurácia de 99% e acertar apenas 60% uma classificação específica (mais sobre aqui);
* **AUC:** fornece uma medida agregada de desempenho em todos os limites de classificação possíveis. Uma maneira de interpretar AUC é como a probabilidade de que o modelo classifique um exemplo positivo aleatório mais alto do que um exemplo negativo aleatório. Quanto mais próximo de 1, melhor. Um modelo cujas previsões estão 100% erradas tem uma AUC de 0,0; aquele cujas previsões são 100% corretas tem uma AUC de 1,0 (mais sobre aqui);
* **Recall:** qual a porcentagem de classificações positivas (em nosso exemplo: 1 - Fraudulentas) o modelo consegue prever. Pode ser resumida em: "Dentre os positivos reais, qual a proporção de acerto?".

Vamos utilizar a **acurácia** como métrica de avaliação, pois é uma boa medida geral para esse problema específico. 

No entanto, caso tivéssemos mais informações sobre o problema, poderíamos considerar utilizar outras métricas para avaliar a performance dos modelos.

### **Tuning de Hiperparametros**

Utilizando o PyCaret, a otimização dos parâmetros é extremamente simples. 

O método `tune_model` permite passar como parâmetro o modelo a ser otimizado e a métrica pela qual você quer otimizar. 

Após a execução, é possível receber os resultados da otimização. 

Além disso, o PyCaret oferece diversas opções para ajustar os parâmetros do modelo, como a alteração da profundidade da árvore, ajuste de parâmetros de regularização, escolha do número de árvores, entre outros. 

Isso permite que você ajuste os parâmetros para obter o melhor resultado possível para o seu problema.

```
# tuning de parâmetros
tuned_lightgbm = tune_model(lightgbm, optimize='Accuracy')
```

A função `tunemodel()` realiza uma pesquisa aleatória de hiperparâmetros para encontrar a melhor configuração. Por padrão, ela é configurada para maximizar o R2, mas isso pode ser alterado usando o parâmetro optimize. 

Por exemplo, *'tunemodel('gbr', optimize = 'MAE')'* irá retornar os parâmetros para obter o menor MAE possível (quanto menor, melhor para MAE).

No nosso projeto específico, vamos preferir usar a acurácia (**Accuracy**) como métrica de avaliação. 

No entanto, para definir a métrica ideal para seu problema específico, é importante ter um bom entendimento do contexto e obter a colaboração das outras pessoas envolvidas no projeto para que a decisão seja tomada de forma não arbitrária.

As métricas não devem ser o único critério para avaliar o desempenho do modelo e escolher o melhor modelo para produção. Outros fatores, como o tempo de treinamento e o desvio padrão dos folds, também devem ser considerados.

Embora o `lightgbm` tenha se saído melhor quando os parâmetros foram configurados de forma padrão, outros modelos, *com hiperparâmetros ajustados*, poderiam ter tido uma melhora significativa e ultrapassado o resultado obtido aqui.

Uma vez que o modelo é finalizado usando a função `finalizemodel()`, todo o dataset, incluindo a parte que havia sido separada para teste anteriormente, é usado para o treino. 

Dessa forma, se o modelo for usado para fazer previsões no conjunto de teste separado pelo PyCaret depois da função `finalizemodel()` ter sido utilizada, os resultados serão enganosos, pois você estará fazendo previsões em dados que já foram vistos pelo modelo.  

Para avaliar isso, vou usar o modelo finalizado com o `predict_model()` para comparar os resultados com os que tivemos na parte anterior.

```
# analisando a mudança nos resultados
predict_model(final_lightgbm);
```

Repare que a acurácia subiu de 0.9303 para 0.9617, apesar do modelo ser o mesmo.
